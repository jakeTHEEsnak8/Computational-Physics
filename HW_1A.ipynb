### Introduction to Vectors and NumPy

Let's first see how to create vectors in Python using the NumPy library. We store the vector components in a NumPy array.  Run the following code cell to execute it by pressing Shift-Enter or clicking the play button.

```python
import numpy as np

# Define 3D vectors using NumPy arrays
A = np.array([1, 2, 3])
B = np.array([-1, 0, 2])

# Display the vectors
print("Vector A = ", A)
print("Vector B = ", B)
```

## Magnitude of a Vector

The magnitude of a vector, also known as its Euclidean norm or "length," is calculated using the Pythagorean theorem. For a 3D vector $ \mathbf{A} = [A_x, A_y, A_z] $, its magnitude is given by:

$$ A=|\mathbf{A}| = \sqrt{A_x^2 + A_y^2 + A_z^2} .$$

In NumPy, you can use NumPy's Linear Algebra module and the `norm()` function to calculate a vector's magnitude. The command looks like this:  `np.linalg.norm()` .

```python
import numpy as np

# Calculate the magnitude of A and B
A_mag = np.linalg.norm(A)
B_mag = np.linalg.norm(B)

# Display the results
print(f"Magnitude of A = {A_mag:.2f}")
print(f"Magnitude of B = {B_mag:.2f}")
```

## Dot Product

The dot product (also known as the scalar product) takes two vectors and returns a single scalar value. It's often used to find the angle between two vectors or to project one vector onto another.

For two vectors $ \mathbf{A} = [A_1, A_2, A_n] $ and $ \mathbf{B} = [B_1, B_2, .B_3] $, the dot product is defined as:

$$ \mathbf{A} \cdot \mathbf{B} = \sum_{i=1}^{n} A_i B_i = A_1 B_1 + A_2 B_2 +A_3 B_3$$

In NumPy, you can use `np.dot(A,B)` or the ` @` operator to write `A @ B`:

```python
import numpy as np

# Finding the dot product between A and B
prod = np.dot(A,B)
print(f"Dot product A and B = {prod}")

# This example shows how to use the @ operator for dot products
prod2 = A @ B
print(f"We get the same result using '@':  A @ B = {prod2}")
```

## Angle Between Two Vectors

The dot product can be written in two ways:
$$ \mathbf{A} \cdot \mathbf{B} = \sum_{i=1}^{n} A_i B_i = A_1 B_1 + A_2 B_2 +A_3 B_3$$
and
$$ \mathbf{A} \cdot \mathbf{B}  = A B \cos \theta. $$

The Python command `np.dot()` uses the first method to calculate the dot product. We can then use the second definiton to solve for the angle $\theta$ once the dot product is known:
$$\theta = \cos^{-1} \left( \frac{ \mathbf{A} \cdot \mathbf{B}}{AB} \right) $$


```python
import numpy as np

# Calculate the magnitude of v1 and v2
A_mag = np.linalg.norm(A)
B_mag = np.linalg.norm(B)

# Calculate the angle of v1 and v2
theta = np.acos(np.dot(A,B) / (A_mag * B_mag))

# Convert to degrees
theta_deg = np.rad2deg(theta)

# Display the results
print(f"Angle between v1 and v2 = {theta_deg:.2f} degrees")
```

$$ \widehat{\mathbf{a}}  = \frac{\mathbf{A}}{A},$$
where we use the "hat" ($\;\widehat{}\;$) to designate a unit vector.



```python
# create a unit vector that points in the direction of vector A
a = A / A_mag
print(f"If A = {A}, the corresponding unit vector is {a}")
```

### Cross Product

The cross product (also known as the vector product) takes two 3D vectors and returns a new 3D vector that is perpendicular to both original vectors. The magnitude of the cross product is related to the area of the parallelogram formed by the two vectors.

For two 3D vectors $ \mathbf{A} = [A_x, A_y, A_z] $ and $ \mathbf{B} = [B_x, B_y, B_z] $, the cross product is defined as:

$ \mathbf{A} \times \mathbf{B} = [ (A_y B_z - A_z B_y), -(A_x B_z - A_z B_x ), (A_x B_y - A_y B_x) ] $

In NumPy, you can use `np.cross(A,B)`.

```python
import numpy as np

# Calculate the cross product C = A x B
C = np.cross(A,B)

print(f"Cross product A x B = {C}")
```
