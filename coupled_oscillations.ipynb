```python
import numpy as np
from scipy.linalg import eigh #python library for linear algebra

# Solve the system of two equal masses and three springs
# connected to two rigid walls
#
# Assume:  m = 1, k = 1

K = np.array([[2, -1],   # set numpy arrays as your matrix
              [-1, 2]])

M = np.array([[1, 0],
              [0, 1]])

eigenvalues, eigenvectors = eigh(K, M)

# extract frequencies of natural modes
omega1 = np.sqrt(eigenvalues[0])
omega2 = np.sqrt(eigenvalues[1])

# extract amplitude vectors of natural modes
mode1 = eigenvectors[:,0] #index slicing
mode2 = eigenvectors[:,1]

# print results
print()
print("mode 1:")
print(f"omega 1 =  {omega1}")
print(f"oscillation amplitudes =  {mode1}")

print()
print("mode 2:")
print(f"omega 1 =  {omega2}")
print(f"oscillation amplitudes =  {mode2}")


```
